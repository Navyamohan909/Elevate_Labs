```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Titanic-Dataset (1).csv")
```

```python
df.info()            # Shows columns, data types, nulls
df.head()            # First 5 rows
df.shape             # Dimensions
df.columns           # Column names
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())       # For numerical columns
print(df.describe(include='object'))  # For categorical columns
```

```python
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
# Calculate and plot the correlation matrix for numerical columns
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']])
plt.show()
```
